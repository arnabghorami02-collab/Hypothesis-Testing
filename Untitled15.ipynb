# ASSIGNMENT HYPOTHESIS
Question 1: What is a null hypothesis (H₀) and why is it important in hypothesis testing?

Sol :     
The null hypothesis (H₀) is a statement that assumes no effect or no difference.

Example: The average time = 10 minutes.

Importance:

It provides a starting point for testing. We use data to decide whether to reject or fail to reject H₀.

Question 2: What does the significance level (α) represent in hypothesis testing?

Sol :      
The significance level (α) is the probability of making a Type I error.

Common value: α = 0.05 (5%)

This means there is a 5% risk of rejecting a true null hypothesis.

Question 3: Differentiate between Type I and Type II errors.

Sol :        

Type I Error --> Rejecting atrue H₀ ( False Positive )

Type II Error --> Not rejecting a false H₀ ( False Negative )

Question 4: Explain the difference between a one-tailed and two-tailed test. Give an example of each.

Sol :      
One-tailed test: Tests in one direction (>, <)

Example: Mean > 10

Two-tailed test: Tests both directions (≠)

Example: Mean ≠ 10

Question 5: A company claims that the average time to resolve a customer complaint is 10 minutes.

A random sample of 9 complaints gives an average time of 12 minutes and a standard deviation of 3 minutes.

At α = 0.05, test the claim.

Sol :      
μ = 10, x̄ = 12, s = 3, n = 9, α = 0.05

Test statistic: t = \frac{\bar{x} - \mu}{s/\sqrt{n}}

Calculation:
t = (12 − 10) / (3/√9) = 2 / 1 = 2

df = 8

Critical t (two-tailed, 0.05) ≈ 2.306

Since 2 < 2.306, we fail to reject H₀

Conclusion: The claim is not rejected.

Question 6: When should you use a Z-test instead of a t-test?

Sol :    

Use Z-test when:

- Population standard deviation (σ) is known
- Sample size is large (n ≥ 30)

Use t-test when:

- σ is unknown
- Sample size is small (n < 30)

Question 7: The productivity of 6 employees was measured before and after a training program.At α = 0.05, test if the training improved productivity.

Sol :     
Differences:  (5, 5, 1, 3, 1, 3)

Mean difference = 3

Standard deviation ≈ 1.79

n = 6

Test statistic: t = \frac{\bar{d}}{s_d/\sqrt{n}}

t ≈ 3 / (1.79/√6) ≈ 4.10

df = 5

Critical t (0.05, one-tailed) ≈ 2.015

Since 4.10 > 2.015, reject H₀

Conclusion: Training improved productivity.

Question 8: A company wants to test if product preference is independent of gender.At α = 0.05, test independence.

Sol :          
Male-A = 20, Male-B = 30

Female-A = 20, Female-B = 30

Test statistic: \chi^2 = \sum \frac{(O - E)^2}{E}

χ² = 5 + 3.33 + 5 + 3.33 = 16.66

df = 1

Critical value (0.05) = 3.84

Since 16.66 > 3.84, reject H₀

Conclusion: Product preference is not independent of gender (they are related).
